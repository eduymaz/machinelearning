```markdown

SELECT

o.order_id,
o.order_date,
c.customer_id,

e.employee_id,
e.birth_date,
e.hire_date,
cat.category_id,
cat.category_name,
sh.shipper_id,
sh.company_name

FROM orders o

JOIN order_details od ON o.order_id = od.order_id
JOIN customers c ON o.customer_id = c.customer_id
JOIN products p ON od.product_id = p.product_id
JOIN categories cat ON p.category_id = cat.category_id
JOIN employees e ON o.employee_id = e.employee_id
JOIN shippers sh ON o.ship_via = sh.shipper_id

```

#### Müşteri Kaç Aydır Bizimle ? 

````python
df["order_date"] = pd.to_datetime(df["order_date"])
df["order_date"]
````

````python
first_orders = df.groupby('customer_id')["order_date"].min().reset_index()
first_orders.rename(columns={'order_date': 'first_order_date'}, inplace=True)
df = df.merge(first_orders, on='customer_id', how='left')

df.head(5)
````

````python
df['customer_sale_months'] = (df['order_date'].dt.year - df['first_order_date'].dt.year) * 12 + (df['order_date'].dt.month -  df['first_order_date'].dt.month)
df
````

#### Müşterinin Toplam Aldığı Ürün Sayısı


````python
product_count = df.groupby('customer_id')['quantity'].sum().reset_index()
product_count.columns = ['customer_id', 'total_product_count']

df = df.merge(product_count, on='customer_id', how='left')

df.head(5)
````

#### Müşterinin Toplam Harcaması

````python
df['effective_price'] = df['unit_price'] * df['quantity'] * (1-df['discount'])
spend = df.groupby('customer_id')['effective_price'].sum().reset_index()
spend.columns = ['customer_id', 'total_spend_amount']

df = df.merge(spend, on='customer_id', how='left')

df.head(5)
````
